# Compute BLEU Score

```python
from nltk.translate.bleu_score import sentence_bleu
```

```python
i=0
gram1=0
while i<len(predicted):
  print(f'Predicted-> {predicted[i]}')
  print(f'Actual-> {groundtruth[i]}')
  # # print(f'Predicted Split form-> {predicted[i].split()}')
  # # print(f'Actual Split form-> {groundtruth[i].split()}')
  print(f'1-Gram-> {sentence_bleu([groundtruth[i].split()],predicted[i].split(),weights=(1.0, 0, 0, 0))}')

  # print('Predict \t \t \t \t Actual \t \t \t 1-Gram')
  # print(f'{predicted[i]:30}        {groundtruth[i]:25}         {sentence_bleu([groundtruth[i].split()],predicted[i].split(),weights=(1.0, 0, 0, 0))}')


  gram1+=sentence_bleu([groundtruth[i].split()],predicted[i].split(),weights=(1.0, 0, 0, 0))
  i=i+1
  print()
```

```python
gram1/len(predicted)
```

# WER

```python
from jiwer import wer
```

```python
i=0
sum=0
while i <len(groundtruth):
  print(f'Groundtruth----> {groundtruth[i]}')
  print(f'Predicted---> {predicted[i]}')
  print(f'Score {wer(groundtruth[i],predicted[i])}')
  sum+=wer(groundtruth[i].split(),predicted[i].split())
  i+=1
```

```python
sum/len(groundtruth)
```
